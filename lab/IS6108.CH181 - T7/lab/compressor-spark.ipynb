```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import mean, col, avg
from pyspark.ml.feature import VectorAssembler, PCA as SparkPCA
from pyspark.ml.clustering import BisectingKMeans
import matplotlib.pyplot as plt

# Tạo Spark session
spark = SparkSession.builder.appName("AnomalyDetection").getOrCreate()

# Đọc file CSV
file_path = "../../../data/Compressor_2022_June-Oct.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)

# Tự động đổi tên các cột có chứa dấu gạch ngang thành dấu gạch dưới
for column in df.columns:
    new_column_name = column.replace("-", "_").replace(".", "_")
    df = df.withColumnRenamed(column, new_column_name)

# Hiển thị tên cột sau khi đổi tên
print("Tên cột sau khi đổi tên:")
print(df.columns)

# Chọn cột đầu tiên làm trục X (Timestamp) và một cột được chọn làm trục Y
timestamp_column = df.columns[0]
selected_column = df.columns[3]  # Thay thế bằng tên cột bạn muốn chọn làm trục Y

# Hàm xử lý dữ liệu thiếu
def handle_missing_data(df, method="mean_fill"):
    if method == "backward_fill":
        df = df.fillna('bfill')
    elif method == "forward_fill":
        df = df.fillna('ffill')
    elif method == "mean_fill":
        df = df.na.fill(df.select([mean(col(c)).alias(c) for c in df.columns[1:]]).first().asDict())
    return df

# Áp dụng phương pháp xử lý mean fill
df = handle_missing_data(df, "mean_fill")

# Hiển thị dữ liệu sau khi xử lý
print("Dữ liệu sau khi xử lý mean fill:")
df.show(5)

# Chọn các cột cần thiết cho mô hình PCA (bỏ qua cột đầu tiên)
assembler = VectorAssembler(inputCols=df.columns[1:], outputCol="features")
df_features = assembler.transform(df)

# Áp dụng PCA để giảm số lượng đặc trưng
pca = SparkPCA(k=2, inputCol="features", outputCol="pca_features")
df_pca = pca.fit(df_features).transform(df_features)

# Sử dụng Bisecting KMeans để phát hiện các nhóm (thay thế cho IsolationForest)
kmeans = BisectingKMeans(k=2, featuresCol="pca_features")
model = kmeans.fit(df_pca)
predictions = model.transform(df_pca)

# Gán nhãn dự đoán (thay thế cột anomaly)
predictions = predictions.withColumnRenamed("prediction", "anomaly")

# Chuyển đổi về Pandas để vẽ biểu đồ
pandas_df = predictions.select(timestamp_column, selected_column, "anomaly").toPandas()

# Vẽ biểu đồ line với trục X là Timestamp và trục Y là selected_column, các điểm bất thường được đánh dấu bằng scatter màu đỏ
plt.figure(figsize=(14, 7))

# Vẽ biểu đồ line cho cột được chọn
plt.plot(pandas_df[timestamp_column], pandas_df[selected_column], label=selected_column, color='blue')

# Vẽ các điểm bất thường bằng scatter màu đỏ
anomalies = pandas_df[pandas_df["anomaly"] == 1]  # Điều chỉnh nếu cần dựa trên nhãn của cụm bất thường
plt.scatter(anomalies[timestamp_column], anomalies[selected_column], color='red', label="Anomalies")

plt.xlabel(timestamp_column)
plt.ylabel(selected_column)
plt.legend()
plt.title("Anomaly Detection using Bisecting KMeans with PCA")
plt.show()
```
