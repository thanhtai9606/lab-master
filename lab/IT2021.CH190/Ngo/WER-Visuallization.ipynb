```python
import matplotlib.pyplot as plt

# Dữ liệu WER
methods = ["Clean Data", "Noise Data", "Kalman", "Spectral", "Weiner"]
wer_values = [22.93, 29.02, 33.48, 37.35, 36.24]

# Tạo biểu đồ
plt.figure(figsize=(8, 5))
plt.bar(methods, wer_values, color=["green", "red", "blue", "purple", "orange"])

# Thêm nhãn và tiêu đề
plt.xlabel("Phương pháp xử lý dữ liệu")
plt.ylabel("Word Error Rate (WER) %")
plt.title("So sánh WER trước và sau khi lọc nhiễu")
plt.ylim(0, max(wer_values) + 5)  # Điều chỉnh giới hạn trục Y để dễ nhìn

# Hiển thị giá trị trên các cột
for i, wer in enumerate(wer_values):
    plt.text(i, wer + 1, f"{wer:.2f}%", ha='center', fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dữ liệu WER
methods = ["Clean Data", "Noise Data", "Kalman + VAD", "Spectral + VAD"]
wer_values = [22.93, 29.02, 57.72, 87.08]

# Vẽ biểu đồ
plt.figure(figsize=(8, 5))
plt.bar(methods, wer_values, color=["green", "red", "blue", "purple"])

# Thêm nhãn và tiêu đề
plt.xlabel("Phương pháp")
plt.ylabel("Word Error Rate (WER) %")
plt.title("So sánh WER sau khi loại bỏ khoảng lặng với VAD")
plt.ylim(0, 100)  # Giới hạn trục Y từ 0 đến 100%

# Hiển thị giá trị trên các cột
for i, wer in enumerate(wer_values):
    plt.text(i, wer + 1, f"{wer:.2f}%", ha='center', fontsize=10)

# Hiển thị biểu đồ
plt.show()
```
